```python
import pandas as pd
from matplotlib import pyplot as plt
import networkx as nx
import seaborn as sns
```

```python
df= pd.read_csv('reddit_worldnews_start_to_2016-11-22.csv',sep=',',nrows=8)
df.plot(kind = 'scatter', x = 'author', y = 'title', color = 'red')
plt.title("Scatter plot")
plt.show()
```

```python
df= pd.read_csv('reddit_worldnews_start_to_2016-11-22.csv',sep=',',nrows=10)
plt.plot(df['title'])
plt.plot(df['author'])
plt.title("Scatter Plot")
plt.xlabel('title')
plt.ylabel('author') 
plt.show()
```

```python
df= pd.read_csv('reddit_worldnews_start_to_2016-11-22.csv',sep=',',nrows=7)
plt.hist(df['author'])
plt.title("Histogram")
plt.show()
```

```python
df= pd.read_csv('reddit_worldnews_start_to_2016-11-22.csv',sep=',',nrows=60)
plt.scatter(df['up_votes'], df['author'])
plt.title("Scatter Plot")
plt.xlabel('up_votes')
plt.ylabel('author')
plt.show()
```

```python
sns.histplot(x='up_votes', data=df, kde=True, hue='author') 
plt.show()
```

```python
df= pd.read_csv('reddit_worldnews_start_to_2016-11-22.csv',sep=',',nrows=15)
plt.bar(df['up_votes'], df['title']) 
plt.title("Bar Chart")
plt.xlabel('up_votes')
plt.ylabel('title')
plt.show()
```
