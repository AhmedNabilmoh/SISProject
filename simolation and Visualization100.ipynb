```python
def initial_state(G):
    state = {}
    for node in G.nodes:
        state[node] = 'life'
    return state
```

```python
initial_state(G)
```

```python
import random

P_AWAKEN = 0.2
def state_transition(G, current_state):
    next_state = {}
    for node in G.nodes:
        if current_state[node] == 'life':
            if random.random() < P_AWAKEN:
                next_state[node] = 'death'
    return next_state
```

```python
test_state = initial_state(G)
state_transition(G, test_state)
```

```python
from simulation import Simulation
sim = Simulation(G, initial_state, state_transition)
```
